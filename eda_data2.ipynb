```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df = pd.read_csv('./train_delete_outlier.csv')
df
```

```python
df.columns
```

```python
df_copy = df.copy()
df_c = df_copy[['연간 소득', '신용 거래 연수', '월 상환 부채액', '최대 신용한도', '현재 미상환 신용액', '개설된 신용계좌 수', '마지막 연체 이후 경과 개월 수', '신용 점수', '현재 대출 잔액']]
```

```python
col_list1 = df_c.columns
len(col_list1)
```

```python
fig, axes = plt.subplots(3, 3, figsize = (20, 20))
sns.boxplot(df_c['연간 소득'], ax = axes[0, 0])
axes[0, 0].set_title('연간 소득')
sns.boxplot(df_c['신용 거래 연수'], ax = axes[0, 1])
axes[0, 1].set_title('신용 거래 연수')
sns.boxplot(df_c['월 상환 부채액'], ax = axes[0, 2])
axes[0, 2].set_title('월 상환 부채액')
sns.boxplot(df_c['최대 신용한도'], ax = axes[1, 0])
axes[1, 0].set_title('최대 신용한도')
sns.boxplot(df_c['현재 미상환 신용액'], ax = axes[1, 1])
axes[1, 1].set_title('현재 미상환 신용액')
sns.boxplot(df_c['개설된 신용계좌 수'], ax = axes[1, 2])
axes[1, 2].set_title('개설된 신용계좌 수')
sns.boxplot(df_c['마지막 연체 이후 경과 개월 수'], ax = axes[2, 0])
axes[2, 0].set_title('마지막 연체 이후 경과 개월 수')
sns.boxplot(df_c['신용 점수'], ax = axes[2, 1])
axes[2, 1].set_title('신용 점수')
sns.boxplot(df_c['현재 대출 잔액'], ax = axes[2, 2])
axes[2, 2].set_title('현재 대출 잔액')
plt.show()
```
